This google colab notebook shows how to load a notebook image from google drive. This was also done using Gemini's AI generative suggestions. It had some struggles, but it wasn't too hard the more I worked with it.

I started first by just giving it the task I was given, which was "load this sat image." It generated four differnt code blocks, which all didn't work for me right away. The main issue was that it didn't really know where the file was, despite me giving it a link to my image. It was able to get close however.

```python
# prompt: Load and display the image found at this link https://drive.google.com/file/d/1-eo1E5XC9Sb-rLNzZjrQM0Bx99MG3Fel/view?usp=drive_link

import os
from google.colab import drive
from IPython.display import Image

# Mount your Google Drive
drive.mount('/content/drive')

# Get the image path
image_path = '/content/drive/My Drive/Colab Notebooks/image.png'

# Check if the image exists
if os.path.exists(image_path):
  # Display the image
  Image(filename=image_path)
else:
  # Print an error message
  print("Image not found")
```

```python
# prompt: load a list of files in this google drive folder "/content/drive/My Drive"

import os

# Get the list of files in the specified Google Drive folder
files = os.listdir("/content/drive/My Drive/Colab Notebooks")

# Print the list of files
print(files)
```
